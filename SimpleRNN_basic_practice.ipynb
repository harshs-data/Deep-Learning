```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, SimpleRNN, Dense
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
import numpy as np # Import numpy

# Sample data
sentences = ["I love this movie", "I hate this film"]
labels = [1, 0]

# Tokenization
tokenizer = Tokenizer()
tokenizer.fit_on_texts(sentences)
sequences = tokenizer.texts_to_sequences(sentences)

# Padding
X = pad_sequences(sequences, maxlen=5)
y = np.array(labels) # Convert y to numpy array

# Model
vocab_size = len(tokenizer.word_index) + 1 # Calculate vocab size
model = Sequential([
    Embedding(input_dim=vocab_size, output_dim=8), # Use vocab_size and remove input_length
    SimpleRNN(16),
    Dense(1, activation='sigmoid')
])

model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

model.fit(X, y, epochs=10, verbose=1)
```

```python
# Get the raw probability
probability = model.predict([pad_data])[0][0]

# Convert to 0 or 1
prediction = 1 if probability > 0.5 else 0

print(f"Probability: {probability}")
print(f"Final Label: {prediction}")
```
