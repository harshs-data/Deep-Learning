```python
import spacy

!python -m spacy download en_core_web_lg
nlp = spacy.load('en_core_web_lg') #loading model
```

```python
spacy.explain('VBG')
```

```python
# function for pos
for word in doc:
  print(word.text, "------->", word.pos_ , spacy.explain(word.tag_))
```

```python
def ner(doc):
  for word in doc:
   print(word.text, "------->", word.pos_ ,'|', spacy.explain(word.tag_))
```

```python
doc2 = nlp('Today is a great day')
ner(doc2)
```

```python
doc3 = nlp('i left the room')
ner(doc3)
```

```python
doc4 = nlp('to the left of the room')
ner(doc4)
```

```python
doc9 = nlp('Will will google campusx')
ner(doc9)
```

```python
# 2. Get the AttributeRuler from the pipeline
ruler = nlp.get_pipe("attribute_ruler")

# 3. Define the pattern
# We look for "Will" at the start of a sentence followed by "will" (lowercase)
patterns = [
    [{"TEXT": "Will", "IS_SENT_START": True}, {"LOWER": "will"}]
]

# 4. Define the correct attributes
# We want to change the first token (index 0) to a Proper Noun
attrs = {"POS": "PROPN", "TAG": "NNP"}

# 5. Add the rule to the ruler
ruler.add(patterns=patterns, attrs=attrs, index=0)
```

```python
doc9 = nlp('Will will google campusx')
ner(doc9)
```
