```python
import spacy
```

```python
nlp = spacy.load('en_core_web_sm')
```

```python
doc = nlp('Apple is looking at buying U.K. startup for $1 billion')
```

```python
doc.text
```

```python
doc[0]
```

```python
# course-grained parts of speech

# parts of speech
doc[2].pos_
```

```python
# fine-grained parts of speech
# additional detailing

doc[2].tag_
```

```python
spacy.explain('VBG')
```

```python
# function for pos
for word in doc:
  print(word.text, "------->", word.pos_ , spacy.explain(word.tag_))
```

```python
def ner(doc):
  for word in doc:
   print(word.text, "------->", word.pos_ ,'|', spacy.explain(word.tag_))
```

```python
doc2 = nlp('Today is a great day')
ner(doc2)
```

```python
doc3 = nlp('i left the room')
ner(doc3)
```

```python
doc4 = nlp('to the left of the room')
ner(doc4)
```
